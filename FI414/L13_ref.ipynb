```python
#@title Download and render PDF pages { display-mode: "form" }
!pip install -q pymupdf

import urllib.request, os
import fitz  # PyMuPDF
from IPython.display import Image, display

PDF_URL  = 'https://raw.githubusercontent.com/yunssamfinance/Teaching/main/old/L13_GrossmanHart1980.pdf'
PDF_PATH = '/content/L13_GrossmanHart1980.pdf'
DPI      = 150  # bump to 200 for sharper text, 100 for faster

urllib.request.urlretrieve(PDF_URL, PDF_PATH)
print('Downloaded:', PDF_PATH, os.path.getsize(PDF_PATH), 'bytes')

doc = fitz.open(PDF_PATH)
print(f'{len(doc)} pages.\n')

for i, page in enumerate(doc, start=1):
    pix = page.get_pixmap(dpi=DPI)
    print(f'--- Page {i} ---')
    display(Image(data=pix.tobytes('png')))

doc.close()
```
